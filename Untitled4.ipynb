```python
import os
import time
import jax
import jax.numpy as jnp
import equinox as eqx
import optax
import distrax
from tqdm import tqdm
import numpy as np
import gym
import gym_super_mario_bros
from gym_super_mario_bros.actions import SIMPLE_MOVEMENT
from nes_py.wrappers import JoypadSpace
from gym.wrappers import GrayScaleObservation, ResizeObservation, FrameStack, RecordVideo
```

```python
class RewardWrapper(gym.Wrapper):
    def __init__(self, env):
        super().__init__(env)
        self.prev_info = {"x_pos": 0, "coins": 0, "flag_get": False}

    def reset(self, **kwargs):
        obs = self.env.reset(**kwargs)
        self.prev_info = {"x_pos": 0, "coins": 0, "flag_get": False}
        return obs

    def step(self, action):
        obs, reward, done, info = self.env.step(action)

        # 在 RewardWrapper 中调整 scale：
        reward += right_movement_reward(info, self.prev_info, scale=0.05)
        reward += jump_reward(action, jump_action=5, bonus=0.2)
        reward += coin_reward(info, self.prev_info, bonus=1.0)
        reward += flag_reward(info, self.prev_info, bonus=50.0)

        reward = np.clip(reward, -10, 10)

        self.prev_info = {
            "x_pos": info.get("x_pos", 0),
            "coins": info.get("coins", 0),
            "flag_get": info.get("flag_get", False),
        }

        return obs, reward, done, info
def make_env():
    env = gym_super_mario_bros.make("SuperMarioBros-v3")
    env = JoypadSpace(env, SIMPLE_MOVEMENT)
    env = RewardWrapper(env)
    env = GrayScaleObservation(env, keep_dim=True)
    env = ResizeObservation(env, shape=(84, 84))
    env = FrameStack(env, 4)
    return env

```

```python
class CNNEncoder(eqx.Module):
    conv1: eqx.nn.Conv2d
    conv2: eqx.nn.Conv2d
    conv3: eqx.nn.Conv2d

    def __init__(self, key):
        k1, k2, k3 = jax.random.split(key, 3)
        self.conv1 = eqx.nn.Conv2d(4, 32, kernel_size=8, stride=4, padding=0, key=k1)
        self.conv2 = eqx.nn.Conv2d(32, 64, kernel_size=4, stride=2, padding=0, key=k2)
        self.conv3 = eqx.nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=0, key=k3)

    def __call__(self, x):  # x: (B, C, H, W)
        def apply_single(xi):  # xi: (C, H, W)
            xi = jax.nn.relu(self.conv1(xi))
            xi = jax.nn.relu(self.conv2(xi))
            xi = jax.nn.relu(self.conv3(xi))
            return xi.reshape(1, -1)
        return jax.vmap(apply_single)(x).squeeze(1)  # → (B, 3136)

class ActorCritic(eqx.Module):
    encoder: CNNEncoder
    pi: eqx.nn.Linear
    vf: eqx.nn.Linear

    def __init__(self, key, num_actions):
        key_enc, key_pi, key_vf = jax.random.split(key, 3)
        encoder = CNNEncoder(key_enc)
        dummy = jnp.ones((1, 4, 84, 84), dtype=jnp.float32)
        feat_dim = int(encoder(dummy).shape[-1])
        self.encoder = encoder
        self.pi = eqx.nn.Linear(feat_dim, num_actions, key=key_pi)
        self.vf = eqx.nn.Linear(feat_dim, 1, key=key_vf)

    def __call__(self, x):
        x = x / 255.0
        x = self.encoder(x)
        logits = jax.vmap(self.pi)(x)
        value = jax.vmap(self.vf)(x).squeeze(-1)
        return logits, value
```

```python
def get_action_and_value(model, obs, rng):
    logits, value = model(obs)
    dist = distrax.Categorical(logits=logits)
    action = dist.sample(seed=rng)
    log_prob = dist.log_prob(action)
    entropy = dist.entropy()
    return action, log_prob, value, entropy

def compute_gae(rewards, values, dones, gamma=0.99, lmbda=0.95):
    T = len(rewards)
    advantages = jnp.zeros_like(rewards)
    gae = 0.0
    for t in reversed(range(T)):
        mask = 1.0 - dones[t]
        delta = rewards[t] + gamma * values[t + 1] * mask - values[t]
        gae = delta + gamma * lmbda * mask * gae
        advantages = advantages.at[t].set(gae)
    returns = advantages + values[:-1]
    advantages = (advantages - advantages.mean()) / (advantages.std() + 1e-8)
    return advantages, returns
```

```python
def rollout(env, policy_apply_fn, rng, T=512, render=False):
    obs = env.reset()
    if isinstance(obs, tuple):  # 支持新 step API 的 (obs, info) 返回
        obs, _ = obs

    # 打印关卡信息
    if hasattr(env.unwrapped, "_world") and hasattr(env.unwrapped, "_stage"):
        print(f"🌍 当前关卡：World {env.unwrapped._world}-{env.unwrapped._stage}")
    else:
        print("🌍 当前关卡信息不可用")

    obs_list, action_list, reward_list, done_list, value_list, logprob_list = [], [], [], [], [], []
    prev_x = 0

    for t in range(T):
        obs_np = np.array(obs)

        # ⬇️ 转换 obs 到 (C, H, W)
        if obs_np.shape == (84, 84, 4):
            obs_jax = jnp.transpose(obs_np, (2, 0, 1))
        elif obs_np.shape == (4, 84, 84, 1):
            obs_jax = jnp.squeeze(obs_np, -1)
        else:
            raise ValueError(f"Unexpected obs shape: {obs_np.shape}")

        obs_jax = obs_jax[None, ...]  # (1, 4, 84, 84)
        logits, value = policy_apply_fn(obs_jax)
        dist = distrax.Categorical(logits=logits)

        action = int(dist.sample(seed=rng).item())  # ✅ 防止 DeprecationWarning
        log_prob = float(dist.log_prob(action).item())

        step_result = env.step(action)
        if len(step_result) == 5:  # new step API
            next_obs, reward, terminated, truncated, info = step_result
            done = terminated or truncated
        else:  # old step API
            next_obs, reward, done, info = step_result

        # ⬇️ 加 shaped 奖励（右移 + 跳跃）
        shaped_bonus, prev_x = shaped_reward(info, prev_x)
        reward += shaped_bonus

        if render:
            env.render()

        obs_list.append(obs_jax.squeeze(0))
        action_list.append(action)
        reward_list.append(reward)
        done_list.append(done)
        value_list.append(value.squeeze())
        logprob_list.append(log_prob)

        obs = next_obs
        if done:
            break

    # ⬇️ Bootstrap 最后一个 value
    obs_np = np.array(obs)
    if obs_np.shape == (84, 84, 4):
        obs_jax = jnp.transpose(obs_np, (2, 0, 1))
    elif obs_np.shape == (4, 84, 84, 1):
        obs_jax = jnp.squeeze(obs_np, -1)
    else:
        raise ValueError(f"Unexpected obs shape: {obs_np.shape}")
    obs_jax = obs_jax[None, ...]
    _, next_value = policy_apply_fn(obs_jax)
    value_list.append(next_value.squeeze())

    print("💡 action =", action, "reward =", reward)

    return {
        "obs": jnp.array(obs_list),         # (T, 4, 84, 84)
        "actions": jnp.array(action_list),
        "rewards": jnp.array(reward_list),
        "dones": jnp.array(done_list),
        "values": jnp.array(value_list),    # (T+1,)
        "log_probs": jnp.array(logprob_list),
    }




def make_loss_fn(static_model, clip_eps=0.2, vf_coef=0.5, ent_coef=1.0):
    # ⬇️ 加了 has_aux=True
    @eqx.filter_value_and_grad(has_aux=True)
    def loss_fn(params, batch):
        model = eqx.combine(params, static_model)
        logits, values = model(batch["obs"])
        dist = distrax.Categorical(logits=logits)

        log_probs = dist.log_prob(batch["actions"])
        entropy = dist.entropy().mean()

        ratios = jnp.exp(log_probs - batch["old_log_probs"])
        adv = batch["advantages"]

        pg_loss = -jnp.mean(jnp.minimum(
            ratios * adv,
            jnp.clip(ratios, 1 - clip_eps, 1 + clip_eps) * adv
        ))

        v_loss = 0.5 * ((batch["returns"] - values) ** 2).mean()
        total_loss = pg_loss + vf_coef * v_loss - ent_coef * entropy

        logs = {
            "pg_loss": pg_loss,
            "v_loss": v_loss,
            "entropy": entropy,
        }

        return total_loss, logs

    return loss_fn



def make_train_step(loss_fn):
    @jax.jit
    def train_step(params, static_model, opt_state, batch, optimizer):
        (loss, logs), grads = loss_fn(params, batch)
        updates, opt_state = optimizer.update(grads, opt_state)
        params = optax.apply_updates(params, updates)
        return params, opt_state, logs
    return train_step

def shaped_reward(info, prev_x, jump_bonus=0.2, right_bonus_scale=0.1):
    """根据 Mario 的右移和跳跃状态给予额外奖励"""
    x_pos = info.get("x_pos", prev_x)  # 如果取不到则保持不变
    delta_x = max(x_pos - prev_x, 0)  # 只奖励右移
    right_bonus = delta_x * right_bonus_scale
    jump_bonus_val = jump_bonus if info.get("status", "") == "jumping" else 0.0
    return right_bonus + jump_bonus_val, x_pos

def right_movement_reward(info, prev_info, scale=0.05):
    dx = info.get("x_pos", 0) - prev_info.get("x_pos", 0)
    return scale * max(0, dx)

def jump_reward(action, jump_action=5, bonus=0.5):
    return bonus if action == jump_action else 0.0

def coin_reward(info, prev_info, bonus=1.0):
    dcoins = info.get("coins", 0) - prev_info.get("coins", 0)
    return bonus * max(0, dcoins)

def flag_reward(info, prev_info, bonus=5.0):
    if info.get("flag_get", False) and not prev_info.get("flag_get", False):
        return bonus
    return 0.0
```

```python
def ppo_train(model, optimizer, opt_state, env, rng, total_steps=50_000,
              batch_size=128, minibatch_size=32, update_epochs=4):
    params, static_model = eqx.partition(model, eqx.is_inexact_array)
    loss_fn = make_loss_fn(static_model)

    @jax.jit
    def train_step(params, opt_state, batch):
        (loss, logs), grads = loss_fn(params, batch)
        updates, new_opt_state = optimizer.update(grads, opt_state, params)
        new_params = optax.apply_updates(params, updates)
        return new_params, new_opt_state, logs

    steps = 0
    rewards_log = []
    pbar = tqdm(total=total_steps)

    while steps < total_steps:
        rng, subkey = jax.random.split(rng)
        batch = rollout(env, eqx.combine(params, static_model), subkey, T=batch_size)

        # ⬇️ 注意这里 values 是 T+1 长度，其余是 T
        values = batch["values"]
        rewards = batch["rewards"]
        dones = batch["dones"]
        advantages, returns = compute_gae(rewards, values, dones)

        actual_size = rewards.shape[0]  # T 步数

        data = {
            "obs": batch["obs"][:actual_size],              # 对齐 T
            "actions": batch["actions"][:actual_size],
            "old_log_probs": batch["log_probs"][:actual_size],
            "advantages": advantages,
            "returns": returns,
        }

        idxs = jnp.arange(actual_size)

        for _ in range(update_epochs):
            rng, perm_key = jax.random.split(rng)
            perm = jax.random.permutation(perm_key, idxs)

            for i in range(0, actual_size, minibatch_size):
                mb_idx = perm[i:i+minibatch_size]
                mb = {k: v[mb_idx] for k, v in data.items()}
                params, opt_state, logs = train_step(params, opt_state, mb)

                print(f"[{steps}] pg={logs['pg_loss']:.4f}, v={logs['v_loss']:.4f}, ent={logs['entropy']:.4f}")

        avg_reward = rewards.mean().item()
        rewards_log.append(avg_reward)
        steps += actual_size

        pbar.update(actual_size)
        pbar.set_description(f"🎯 reward={avg_reward:.2f} | steps={steps}")

    pbar.close()
    return eqx.combine(params, static_model), rewards_log
```

```python
def record_video(model, env_fn, filename="ppo_mario.mp4", max_steps=1000):
    os.makedirs("videos", exist_ok=True)
    env = RecordVideo(env_fn(), video_folder="videos", name_prefix="ppo_mario", episode_trigger=lambda x: True)
    obs = env.reset()
    obs_np = np.array(obs)
    if obs_np.ndim == 3 and obs_np.shape[-1] == 4:
        obs_jax = jnp.transpose(obs_np, (2, 0, 1))  # (84, 84, 4) → (4, 84, 84)
    elif obs_np.ndim == 4 and obs_np.shape[-1] == 1:
        obs_jax = jnp.squeeze(obs_np, -1)           # (4, 84, 84, 1) → (4, 84, 84)
    else:
        raise ValueError(f"Unexpected obs shape: {obs_np.shape}")
    obs = obs_jax[None, ...]

    step = 0
    done = False
    while not done and step < max_steps:
        logits, _ = model(obs)
        action = jnp.argmax(logits, axis=-1)
        obs_, reward, done, _ = env.step(int(action[0]))

        obs_np = np.array(obs_)
        if obs_np.ndim == 3 and obs_np.shape[-1] == 4:
            obs_jax = jnp.transpose(obs_np, (2, 0, 1))
        elif obs_np.ndim == 4 and obs_np.shape[-1] == 1:
            obs_jax = jnp.squeeze(obs_np, -1)
        else:
            raise ValueError(f"Unexpected obs shape: {obs_np.shape}")
        obs = obs_jax[None, ...]

        step += 1

    env.close()
    print("🎥 Saved video to ./videos/")
```

```python
env = make_env()
num_actions = env.action_space.n
rng = jax.random.PRNGKey(0)

model = ActorCritic(jax.random.PRNGKey(42), num_actions)
optimizer = optax.chain(optax.clip_by_global_norm(1.0), optax.adam(3e-4))
opt_state = optimizer.init(eqx.filter(model, eqx.is_inexact_array))

trained_model, logs = ppo_train(model, optimizer, opt_state, env, rng)
record_video(trained_model, make_env)
```
